```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img = cv2.imread('Wallpaper.jpg')
```

```python
plt.imshow(img)
```

```python
crop_img = img[0:550, 400:1200]
```

```python
plt.imshow(crop_img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('Wallpaper.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

```python
img_rgb.shape
```

```python
img_gray.shape
```

```python
img_s1 = cv2.resize(img_rgb, (100, 50))
plt.imshow(img_s1)
```

```python
#RESIZE BY RATIO

w_r = 0.25
h_r = 0.25

img_ratio = cv2.resize(img_rgb, (0, 0), img_rgb, w_r, h_r)
plt.imshow(img_ratio)
```

```python
# FLIPPING

flp1 = cv2.flip(img, 0)
plt.imshow(flp1)
```

```python
flp2 = cv2.flip(img, 1)
plt.imshow(flp2)
```

```python
flp = cv2.flip(img, -1)
plt.imshow(flp)
```

```python
cv2.imwrite('flip_img_1.jpeg', flp)
```

```python
# FIGURE SIZE

fig = plt.figure(figsize = (10, 50))
plt.imshow(img_gray)
```

```python
# applying threshold to convert grayscale to binary

ret, thresh = cv2.threshold(img_gray, 100, 255, 0)
# cv2.imshow('Binary Image', thresh)
plt.imshow(thresh, cmap = 'gray')
```

```python
# XOR OPERATION

# a1 = cv2.cvtColor("img_1.jpeg", cv2.COLOR_BGR2GRAY)
# a2 = cv2.cvtColor("img_2.jpeg", cv2.COLOR_BGR2GRAY)

res_xor = cv2.bitwise_xor(flp1, flp2, mask = None)
res_or = cv2.bitwise_or(flp1, flp2, mask = None)
res_and = cv2.bitwise_and(flp1, flp2, mask = None)

plt.imshow(res_xor, cmap = 'gray')
plt.imshow(res_or, cmap = 'gray')
plt.imshow(res_and, cmap = 'gray')
```

```python
weighted_sum = cv2.addWeighted(flp1, 0.5, flp2, 0.4, 0)
plt.imshow(weighted_sum)
```

```python
weighted_sub = cv2.subtract(flp1, flp2)
plt.imshow(weighted_sub)
```

```python
arr = np.zeros((100, 100))
plt.imshow(arr, cmap = 'gray')
```

# Q1) What do you mean by gray level?
  A1) The minimum grey level is 0. The maximum grey level depends on the digitisation depth of the image. For an 8-bit-deep image it is 255. 
     In a binary image a pixel can only take on either the value 0 or the value 255. In contrast, in a greyscale or colour image a pixel can 
     take on any value between 0 and 255.

# Q4) Specify the elements of the DIP system
  A4)1. Image acquisition: to acquire a digital image 
     2. Image preprocessing: to improve the image in ways that increase the chances for success of the other processes. 
     3. Image segmentation: to partitions an input image into its constituent parts or objects. 
     4. Image representation: to convert the input data to a form suitable for computer processing
     5. Image description: to extract features that result in some quantitative information of interest or features that are basic for 
        differentiating one class of objects from another. 
     6. Image recognition: to assign a label to an object based on the information provided by its descriptors.
     7. Image interpretation: to assign meaning to an ensemble of recognized objects.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img_gray = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap = 'gray')
```

```python
print(img_gray.shape)
```

```python
gauss_noise = np.zeros((1080, 1920), dtype = np.uint8)
cv2.randn(gauss_noise,128, 20)
gauss_noise = (gauss_noise*0.5).astype(np.uint8)
plt.imshow(gauss_noise, cmap = 'gray')
```

```python
gn_img = cv2.add(img_gray, gauss_noise)
plt.imshow(gn_img)
```

```python
fig = plt.figure(dpi = 300)

fig.add_subplot(1,3,1)
plt.imshow(img_gray, cmap = 'gray')
plt.axis('off')
plt.title('original')

fig.add_subplot(1,3,2)
plt.imshow(gauss_noise, cmap = 'gray')
plt.axis('off')
plt.title('Gaussian Noise')

fig.add_subplot(1,3,3)
plt.imshow(gn_img, cmap = 'gray')
plt.axis('off')
plt.title('Combined')
```

```python
#Impulse Noise (salt and pepper)

img_noise = np.zeros((1080, 1920), dtype = np.uint8)
cv2.randn(img_noise, 0,255)
img_noise = cv2.threshold(img_noise, 245, 255, cv2.THRESH_BINARY)[1]
```

```python
in_img = cv2.add(img_gray, img_noise)
```

```python
fig = plt.figure(dpi = 300)

fig.add_subplot(1,3,1)
plt.imshow(img_gray, cmap = 'gray')
plt.axis('off')
plt.title('Original')

fig.add_subplot(1,3,2)
plt.imshow(img_noise, cmap = 'gray')
plt.axis('off')
plt.title('Impulse Noise')


fig.add_subplot(1,3,3)
plt.imshow(in_img, cmap = 'gray')
plt.axis('off')
plt.title('Combined')
```

```python
#blur image

blur = cv2.blur(in_img, (9,9))

plt.subplot(121), plt.imshow(in_img, cmap = 'gray'), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blur, cmap = 'gray'), plt.title('Blurred')
plt.xticks([]), plt.yticks([])
```

```python
img_gsn3 = cv2.GaussianBlur(in_img,(3,3),cv2.BORDER_DEFAULT)
img_gsn9 = cv2.GaussianBlur(in_img,(9,9),cv2.BORDER_DEFAULT)
img_gsn15 = cv2.GaussianBlur(in_img,(15,15),cv2.BORDER_DEFAULT)
```

```python
fig = plt.figure(dpi = 300)

plt.subplot(141),plt.imshow(in_img, cmap='gray'),plt.title('Noisy Original')
plt.xticks([]),plt.yticks([])
plt.subplot(142),plt.imshow(img_gsn3, cmap='gray'),plt.title('Gkernal 3x3')
plt.xticks([]),plt.yticks([])
plt.subplot(143),plt.imshow(img_gsn9, cmap='gray'),plt.title('Gkernal 9x9')
plt.xticks([]),plt.yticks([])
plt.subplot(144),plt.imshow(img_gsn15, cmap='gray'),plt.title('Gkernal 15x15')
plt.xticks([]),plt.yticks([])
```

```python
median3 = cv2.medianBlur(in_img, 3)
plt.imshow(median3, cmap = 'gray')

median5 = cv2.medianBlur(in_img, 5)
median7 = cv2.medianBlur(in_img, 7)
median15 = cv2.medianBlur(in_img, 15)
```

```python
fig = plt.figure(dpi = 300)

plt.subplot(151),plt.imshow(in_img, cmap='gray'),plt.title('Noisy Original')
plt.xticks([]),plt.yticks([])
plt.subplot(152),plt.imshow(median3, cmap='gray'),plt.title('M 3x3')
plt.xticks([]),plt.yticks([])
plt.subplot(153),plt.imshow(median5, cmap='gray'),plt.title('M 5x5')
plt.xticks([]),plt.yticks([])
plt.subplot(154),plt.imshow(median7, cmap='gray'),plt.title('M 7x7')
plt.xticks([]),plt.yticks([])
plt.subplot(155),plt.imshow(median15, cmap='gray'),plt.title('M 15x15')
plt.xticks([]),plt.yticks([])
```

**1. What is point processing?**

Ans. Point processing is a type of digital image processing where the processing is done on a pixel-by-pixel basis without taking into account the relationship between neighboring pixels. It involves transforming the intensity values of individual pixels in an image without changing their position.

**2. Consider an input image row [4 3 2 1] with intensity values in the range of 0 to 15.
Determine the negative of the image row.**

Ans. To obtain the negative of the image row [4 3 2 1], we can subtract each pixel value from 15 as follows:

Negative of [4 3 2 1] = [15-4 15-3 15-2 15-1] = [11 12 13 14]

Therefore, the negative of the input image row [4 3 2 1] is [11 12 13 14].

**3. Determine the new value of the central pixel of the following image by applying 3x3 size:**
**a) Mean filter b) Median filter c) Mode filter**

**10 11 11**

**10 255 11**

**12 12 11**

Ans. *a) Mean filter:*

The mean filter replaces the central pixel value with the average value of the 3x3 neighborhood. So, the new value of the central pixel after applying the mean filter is:

(10+11+11+10+255+11+12+12+11)/9 = 38

Therefore, the new value of the central pixel after applying the mean filter is 83.

*b) Median filter:*

The median filter replaces the central pixel value with the median value of the 3x3 neighborhood.

10, 10, 11, 11, *11*, 11, 11, 12, 12, 255

The median value is the middle value when the values are sorted. In this case, the median value is 11.

Therefore, the new value of the central pixel after applying the median filter is 11.

*c) Mode filter:*

The mode filter replaces the central pixel value with the mode value of the 3x3 neighborhood, which is the most frequently occurring value. In this case, the value 11 occurs most frequently in the neighborhood.

Therefore, the new value of the central pixel after applying the mode filter is 11.
